#### EDA VARIABLES CATEGÓRICAS

```python
def graficos_eda_categoricos(cat):

    #Calculamos el número de filas que necesitamos
    from math import ceil
    #ceil nos permite redondear UP
    filas = ceil(cat.shape[1] / 2)

    #Definimos el frame del grafico
    f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))

    #Aplanamos para iterar por el gráfico como si fuera de 1 dimensión en lugar
    # de 2
    ax = ax.flat

    #Añadimos un bucle que añade gráficos de barras para ilustrar la
    # distribucion de cada variable categorica.
    for muestras, variable in enumerate(cat):
        cat[variable].value_counts().plot.barh(ax = ax[muestras])
        ax[muestras].set_title(variable, fontsize = 12, fontweight = "bold")
        ax[muestras].tick_params(labelsize = 12)
```

```python
graficos_eda_categoricos(df.select_dtypes('O'))
```

Interesante observar que el puesto donde mayor porcentaje de dimisiones es Sales Executive, seguido Research Scientist y Laboratory Technician con un amplio margen respecto los demás puestos.

```python
df[df.puesto == 'Sales Executive'].puesto.count()/df.puesto.count()*100
```

Conclusiones:

* mayor_edad solo tiene un valor (constante) eliminamos.
* Sobre las imputaciones pendientes de variables categóricas, al ser pocos los valores nulos, imputaremos el valor más común en cada feature.
  * educacion: imputar por 'Universitaria'
  * satisfaccion_trabajo: imputar por 'Alta'
  * implicacion: imputar por 'Alta'

```python
df.drop(columns = 'mayor_edad', inplace = True)

df['educacion'] = df['educacion'].fillna('Universitaria')

df['satisfaccion_trabajo'] = df['satisfaccion_trabajo'].fillna('Alta')

df['implicacion'] = df['implicacion'].fillna('Alta')
```

#### EDA VARIABLES NUMÉRICAS

```python
def estadisticos_cont(num):
    #Calculamos describe
    estadisticos = num.describe().T
    #Añadimos la mediana
    estadisticos['median'] = num.median()
    #Reordenamos para que la mediana esté al lado de la media
    estadisticos = estadisticos.iloc[:,[0,1,8,2,3,4,5,6,7]]
    #Lo devolvemos
    return(estadisticos)
```

```python
estadisticos_cont(df.select_dtypes('number'))
```

Conclusiones:

* Empleados solo tiene un valor así que no aporta.
* Sexo tiene 4 valores y no se especifica nada en la descripción del dataser. Borramos por si se tratase de algún error.
* Horas quincena solo tiene una valor así que no aporta.
* De los nulos pendientes de imputación que sean numéricas solo está el sexo, pero como la vamos a eliminar ya no hay que imputar nada.

### ¿Cual es el impacto económico de este problema?

Según el estudio "Cost of Turnover" del Center for American Progress [(There Are Significant Business Costs to Replacing Employees)](https://www.americanprogress.org/article/there-are-significant-business-costs-to-replacing-employees/):

* El coste de la fuga de los empleados que ganan menos de 30000 es del 16,1% de su salario

* El coste de la fuga de los empleados que ganan entre 30000-50000 es del 19,7% de su salario

* El coste de la fuga de los empleados que ganan entre 50000-75000 es del 20,4% de su salario

* El coste de la fuga de los empleados que ganan más de 75000 es del 21% de su salario

```python
# Creamos una nueva variable salario_ano del empleado
df['salario_ano'] = df.salario_mes.transform(lambda x: x*12)
df[['salario_mes','salario_ano']]
```

```python
# Calculamos el impacto económico de cada empleado si deja la empresa

#Lista de condiciones
condiciones = [(df['salario_ano'] <= 30000),
               (df['salario_ano'] > 30000) & (df['salario_ano'] <= 50000),
               (df['salario_ano'] > 50000) & (df['salario_ano'] <= 75000),
               (df['salario_ano'] > 75000)]

#Lista de resultados
resultados = [df.salario_ano * 0.161, df.salario_ano * 0.197, df.salario_ano * 0.204, df.salario_ano * 0.21]

#Aplicamos select para aplicar rapidamente los resultados basados en las condiciones al dataset.
df['impacto_abandono'] = np.select(condiciones,resultados, default = -999)

df
```

### Podemos seguir trazando estrategias asociadas a los insights de abandono:

Según lo visto anteriormente, el 22.18% de las dimisiones anuales de la empresa, son representantes de ventas. ¿Tendría sentido hacer un plan específico para ellos? ¿Cual sería el coste ahorrado si disminuimos la fuga un 30%?

Primero vamos a calcular el % de representantes de ventas que se han ido el año pasado.

```python
total_repre_pasado = len(df.loc[df.puesto == 'Sales Representative'])
abandonos_repre_pasado = len(df.loc[(df.puesto == 'Sales Representative') & (df.abandono == 1)])
porc_pasado = abandonos_repre_pasado / total_repre_pasado * 100

print(f'{round(porc_pasado, 2)}%')
```

Ahora vamos a estimar cuántos se irian este año.

```python
total_repre_actual = len(df.loc[(df.puesto == 'Sales Representative') & (df.abandono == 0)])
se_iran = int(total_repre_actual * porc_pasado)

se_iran
```

Sobre ellos cuantos podemos retener (hipótesis 30%) y cuanto dinero puede suponer

```python
retenemos = int(se_iran * 0.3)

ahorramos = df.loc[(df.puesto == 'Sales Representative') & (df.abandono == 0),'impacto_abandono'].sum() * porc_pasado * 0.3

print(f'Podemos retener {retenemos} representantes de ventas y ello supondría ahorrar {round(ahorramos, 2)}$.')
```

Este dato también es muy interesante porque nos permite determinar el presupuesto para acciones de retención por departamento o perfil.

Ya que sabemos que podemos gastarnos hasta 37.000$ sólo en acciones específicas para retener a representantes de ventas y se estarían pagando sólas con la pérdida evitada

## MODELO DE PREDICCIÓN

```python
# Copiamos el dataset para no pisar los datos de partida.
df_ml = df.copy()
```

```python
df_ml.info()
```

### PREPARACIÓN DE LOS DATOS PARA LA MODELIZACIÓN

#### Transformar todas las variables categóricas a númericas

Vamos a hacer OneHotEncoding directamente ya que el dataset tiene pocas variables categóricas y pocas features en general. Además, OHE suele ofrecer mejores resultados.

```python
from sklearn.preprocessing import OneHotEncoder

#Categóricas (object)
cat = df_ml.select_dtypes('O')

#Instanciamos el encoder
ohe = OneHotEncoder(sparse = False)

#Entrenamos
ohe.fit(cat)

#Aplicamos
cat_ohe = ohe.transform(cat)

#Ponemos los nombres
cat_ohe = pd.DataFrame(cat_ohe, columns = ohe.get_feature_names_out(input_features = cat.columns)).reset_index(drop = True)
```

```python
cat_ohe
```

#### Dataframe final

Seleccionamos las variables numéricas para poder juntarlas a las cat_hoe

```python
num = df.select_dtypes('number').reset_index(drop = True)
```

Las juntamos todas en el dataframe final

```python
df_ml = pd.concat([cat_ohe,num], axis = 1)
df_ml
```

### DISEÑO DEL MODELO

#### Separación features y target

```python
x = df_ml.drop(columns='abandono')
y = df_ml['abandono']
```

#### Separación train y test

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.3)
```

### ENTRENAMIENTO DEL MODELO

Para este caso vamos a utilizar un DecisionTreeClassifier ya que como vimos en el EDA, es facil observar un perfil tipo de empleado que dimite y además, siendo esto un problema típico que se podría encontrar alguien de RRHH, con la grafica del arbol de decisión, veremos de forma clara el camino a seguir para detectar un posible empleado que se va a ir.


```python
from sklearn.tree import DecisionTreeClassifier

#Instanciar
ac = DecisionTreeClassifier(max_depth=4)

#Entrenar
ac.fit(train_x,train_y)
```

### PREDICCIÓN Y VALIDACIÓN SOBRE TEST

```python
# Predicción
pred = ac.predict_proba(test_x)[:, 1]
pred[:20]
```

Analizamos los resultados con la métrica Area Under the Receiver Operating Characteristic Curve (ROC AUC). Para esta metrica, el clasificador perfecto seria 1 y el peor 0.

```python
# Evaluación
from sklearn.metrics import roc_auc_score

roc_auc_score(test_y,pred)
```

En este caso podemos concluir que el resultado es bueno.
De cara a mejorar la efectividad del modelo de podrían buscar muestras extra o cruzar este dataset con features extra que fuesen de interés.

### INTERPRETACIÓN

#### Diagrama del árbol

```python
from sklearn.tree import plot_tree

plt.figure(figsize = (50,50))

plot_tree(ac,
          feature_names= test_x.columns,
          impurity = False,
          node_ids = True,
          proportion = True,
          rounded = True,
          precision = 2);
```

Este tipo de visualización del árbol nos permite comprender de forma rápida y amigable la lógica del modelo.

#### Importancia de las variables

```python
pd.Series(ac.feature_importances_,index = test_x.columns).sort_values(ascending = False).plot(kind = 'bar', figsize = (30,20));
```

Observando la importancia para el modelo calculada por este mismo, nos podemos hacer una idea de que las doce primeras variables mostradas en el gráfico están aportando el grueso de la efectividad del modelo de predicción. Por lo tanto, si la adquisición o guardado de los datos correspondientes al resto de las variables, podría considerarse reducir el tamaño del modelo y así ahorrar tambien en costes procesado de datos por un lado y por el otro, reducir el tiempo de inferencia del modelo (aunque en este caso sea muy bajo).
